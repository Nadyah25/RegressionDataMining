```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

dts_0154 = pd.read_csv('Dataset_0154.csv')
Weeks_0154 = dts_0154.iloc[:, :-1].values 
Savings_2816 = dts_0154.iloc[:, 1].values 
dts_0154

X_TRAIN = Weeks_0154
y_TRAIN = Savings_2816

linreg_0154 = LinearRegression()
linreg_0154.fit(X_TRAIN, y_TRAIN)

Dody_0154 = plt
Dody_0154.scatter(X_TRAIN, y_TRAIN, color='green')
Dody_0154.plot(X_TRAIN, linreg_0154.predict(X_TRAIN), color='black')
Dody_0154.title('Prediction')
Dody_0154.xlabel('Weeks')
Dody_0154.ylabel('Savings')
Dody_0154.show()

# Prediction result of saving 3 weeks
savings_predict = linreg_0154.predict(np.array([3]).reshape(1, 1))
print("You can saves of $", savings_predict, "to your wallet!")
```
